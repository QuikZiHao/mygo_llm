```python
import os
import pandas as pd

# Define the directory containing the CSV files
csv_dir = r'assets/result/csv'

# List all CSV files in the directory
csv_files = [os.path.join(csv_dir, f) for f in os.listdir(csv_dir) if f.endswith('.csv')]

# Read and merge all CSV files
dataframes = [pd.read_csv(file) for file in csv_files]
merged_df = pd.concat(dataframes, ignore_index=True)
merged_df
```

```python
have_chat_data = merged_df[merged_df["have_sub"]]
have_chat_data
```

```python
grouped = have_chat_data.groupby(['dir_name', 'image_name'])['speech'].nunique()
multiple_speech = grouped[grouped > 1]
count = multiple_speech.count()
count
```

```python
grouped = have_chat_data.groupby('speech').size()
same_speech = grouped[grouped > 1]
count = same_speech.count()
count
```

## Vanish the same speech

```python
have_chat_data_sorted = have_chat_data.sort_values(by=['speech', 'score'], ascending=[True, False])
have_chat_data = have_chat_data_sorted.drop_duplicates(subset='speech', keep='first')
have_chat_data
```

```python
have_chat_data.sort_values(by='score', ascending=False)
```

```python
low_score_count = have_chat_data[have_chat_data['score'] < 0.6].shape[0]
low_score_count
```

```python
have_chat_data = have_chat_data[have_chat_data['score'] > 0.6]
have_chat_data
```

```python
have_chat_data['frame_number'] = have_chat_data['image_name'].apply(lambda x: int(x.split('_')[1].split('.')[0]))
have_chat_data['dir_number'] = have_chat_data['dir_name'].apply(lambda x: int(x.split('\\')[-1]))

sorted_df = have_chat_data.sort_values(by=['dir_number', 'frame_number'])

sorted_df
```

```python
sorted_df.to_csv(r"assets/speech/cleaned_data.csv", encoding='utf-8', index=False)
```
