# ASS

```python
import os
import pandas as pd

# Define the directory containing the CSV files
csv_dir = r'assets/subtitle_result/csv'

# List all CSV files in the directory
csv_files = [os.path.join(csv_dir, f) for f in os.listdir(csv_dir) if f.endswith('.csv')]

# Read and merge all CSV files
dataframes = []

for file in csv_files:
    df = pd.read_csv(file)
    # Add a new column to track the source file
    df['eps'] = os.path.basename(file).split('.')[0]
    dataframes.append(df)

# Concatenate all DataFrames
merged_df = pd.concat(dataframes, ignore_index=True)

# Display the merged DataFrame
merged_df
```

```python
delete_speech = "m 20 0 b 10 0 0 10 0 20 b 0 31 10 40 20 40 b 31 40 40 31 40 20 b 40 10 31 0 20 0"

df_cleaned = merged_df[merged_df['speech'] != delete_speech]
df_cleaned
```

```python
from langdetect import detect, DetectorFactory

# Set the seed to ensure consistent results
DetectorFactory.seed = 0

# Define regular expressions to detect Chinese, Japanese, and English text
def detect_language(text: str) -> str:
    text = str(text)
    try:
        # Use langdetect to detect the language
        language = detect(text)
        
        # Map the detected language to your custom categories
        if language == 'zh-cn' or language == 'zh-tw':
            return "Chinese"
        elif language == 'ja':
            return "Japanese"
        elif language == 'en':
            return "English"
        else:
            return "Unknown"
    except Exception as e:
        # If an error occurs, return Unknown
        return "Unknown"

# Example: Applying this to your DataFrame
df_cleaned['language'] = df_cleaned['speech'].apply(detect_language)

df_cleaned
```

```python
df_chinese = df_cleaned[df_cleaned['language'].isin(['Chinese', 'English', 'Unknown'])]
df_japanese = df_cleaned[df_cleaned['language'].isin(['Japanese', 'English', 'Unknown'])]
```

```python
import pandas as pd

# 创建 DataFrame
df = df_japanese
result = []
temp_group = []  # 用于暂时存储相同 speech 的连续行

# 遍历 DataFrame
for i, row in df.iterrows():
    if not temp_group:
        temp_group.append(row)
    elif row["speech"] == temp_group[-1]["speech"]:  # 如果 speech 相同，继续合并
        temp_group.append(row)
    else:  # 如果 speech 不同，处理当前 group
        # 获取 min start_time 和 max end_time，合并成一行
        merged_row = {
            "speech": temp_group[0]["speech"],
            "start_time": min([r["start_time"] for r in temp_group]),
            "end_time": max([r["end_time"] for r in temp_group]),
            "start_frame": min([r["start_frame"] for r in temp_group]),  # 处理 start_frame
            "end_frame": max([r["end_frame"] for r in temp_group]),  # 处理 end_frame
            "eps": int(temp_group[0]["eps"])
        }
        result.append(merged_row)
        temp_group = [row]  # 重置为当前行继续合并

# 处理最后一个 group
if temp_group:
    merged_row = {
        "speech": temp_group[0]["speech"],
        "start_time": min([r["start_time"] for r in temp_group]),
        "end_time": max([r["end_time"] for r in temp_group]),
        "start_frame": min([r["start_frame"] for r in temp_group]),
        "end_frame": max([r["end_frame"] for r in temp_group]),
        "eps": int(temp_group[0]["eps"])
    }
    result.append(merged_row)

# 创建 DataFrame 保存最终结果
merged_df = pd.DataFrame(result)
```

```python
merged_df = merged_df.sort_values(by=['eps','start_time'], ascending=[True,True])
```

```python
merged_df = merged_df[merged_df['speech'].str.len() > 1]

merged_df
```

```python
def get_average_frame(row):
    # Get the average of start and end frame
    avg_frame = int((row['start_frame'] + row['end_frame']) / 2)
    avg_frame = (avg_frame // 12) * 12 + 12
    # Make the result a multiple of 12
    return f"frame_{avg_frame:06d}.jpg"# Rounds up to the nearest multiple of 12

def get_start_frame(frame):
    frame_path = (frame // 12) * 12 + 12
    return f"frame_{frame_path:06d}.jpg"

def get_end_frame(frame):
    frame_path = (frame // 12) * 12
    return f"frame_{frame_path:06d}.jpg"

# Apply the adjustment to the 'start_frame' and 'end_frame' columns
merged_df['get_frame'] = merged_df.apply(get_average_frame, axis=1)
merged_df['start_frame'] = merged_df['start_frame'].apply(get_start_frame)
merged_df['end_frame'] = merged_df['end_frame'].apply(get_end_frame)

# Save the result to an Excel file
merged_df.to_excel(r"assets\speech\ass_result_jp.xlsx", index=False)
```
